```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('compresseddata.csv')

```

```python
df.head()
```

```python
df.columns
```

# checking the missing values

```python
print(df.isnull().sum())  # Check for missing values
```

```python
df.info()  # Get a concise summary of the DataFrame
```

```python
df['last review'] =  pd.to_datetime(df['last review'], errors='coerce')  # Convert 'last review' to datetime
```

```python
df.info()  # Get a concise summary of the DataFrame
```

```python
df.fillna({'reviews per month': 0, 'last review' : df['last review'].mean()}, inplace=True)  # Fill missing values in 'last review' with mean
```

```python
df.dropna(subset = ['NAME', 'host name'] , inplace=True)  # Drop rows where 'Name' or 'host name' is NaN
```

```python
df.isnull().sum()  # Check for missing values again after filling and dropping
```

```python
df = df.drop(columns=['license', 'house_rules'], errors='ignore')  # Drop 'license' and 'house rules' columns if they exist
```

```python
df.head()
```

```python
# Remove dollar signs and convert 'price' and 'service fee' to float
# Remove dollar signs and convert to float
df['price'] = df['price' ].replace('[\$, ]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$, ]', '', regex=True).astype(float)
```

```python
df.head()
```

# Remove duplicates

```python
df.drop_duplicates(inplace=True)  # Remove duplicate rows
```

```python
df.info()
```

# Discriptive statistics

```python
df.describe()
```

# Visualization

# What is the distribution of listing prices?

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['price'], bins=50, kde=True , color='red')
plt.title('Distribution of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# How are different room types distributed?

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='room type', color='hotpink')
plt.title('Distribution of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

# How are listing distributed across different neighborhoods?

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, y='neighbourhood group', color='lightgreen' , order=df['neighbourhood group'].value_counts().index)
plt.title('Distribution of Neighbourhood Groups')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

# What is the relationship between price and room type?

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='room type', y='price', hue='room type', data=df, palette='Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price ($)')
plt.legend(title='Room Type')
plt.show()
```

# How has the number of reviews changed over time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()

plt.Figure(figsize=(8,6))
reviews_over_time.plot(kind='line', color = 'red')
plt.title('Number of reviwes over time')
plt.xlabel('Date')
plt.ylabel('Number of Reviwes')
plt.show()
```
